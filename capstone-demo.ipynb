<h2>Psychosocial Predictors of Wellness in College Students</h2>

Specifically, students who possess a greater ability to trust others, take initiative in tasks, behave industriously, and hold a greater sense of identity increase their chances of achieving higher overall wellness.

```python
import matplotlib.pyplot as plt

# Set this variable to True to run an interactive simulator. This will not work on CoCalc.
interactive = False  
#comment out 'matplotlib inline' if using the GUI
%matplotlib inline 

if interactive:
    matplotlib.use('TkAgg')
import pylab
import numpy as np
import networkx as nx
from collections import Counter


class Wellness:
    
    def __init__(self, alpha=0.2, beta=0.2, interactive=False):
    
        self.interactive = interactive
        
        #social network graph - karate club graph
        self.g = nx.karate_club_graph()
        self.g.pos = nx.spring_layout(self.g)
        
        #parameter weights
        self.alpha = alpha #confidence weight
        self.beta = beta #social weight
        
        for i in self.g.nodes:
            self.g.nodes[i]['wellness'] = np.random.normal(0.5,0.1,1)[0]
            self.g.nodes[i]['confidence'] = np.random.uniform()
            self.g.nodes[i]['social'] = np.random.normal(0.5,0.1,1)[0]
        
    def initialize(self):

        self.g = nx.karate_club_graph()
        self.g.pos = nx.spring_layout(self.g)
        
        for i in self.g.nodes:
            self.g.nodes[i]['wellness'] = np.random.normal(0.5,0.1,1)[0]
            self.g.nodes[i]['confidence'] = np.random.uniform()
            self.g.nodes[i]['social'] = np.random.normal(0.5,0.1,1)[0]

    def observe(self):
        
        if self.interactive:
            pylab.cla()  # Clear the current figure
        else:
            pylab.figure()  # Create a new figure

        nx.draw(self.g, vmin = 0, vmax = 1,
                node_color = [self.g.nodes[i]['wellness'] for i in self.g.nodes],
                cmap="viridis",
                node_size = 1000,
                pos = self.g.pos,
                labels = {k: round(v,2) for k, v in nx.get_node_attributes(self.g, 'wellness').items()})
        
        if not self.interactive:
            pylab.show()

    def update(self):
        
        person = np.random.choice(list(self.g))
        neighbors = list(self.g.neighbors(person))
        
        for neighbor in neighbors:
            #if a neighbor is social, we add a small boost to the individual's own social score
            if self.g.nodes[neighbor]['social'] > 0.5:
                self.g.nodes[person]['social'] += 0.015*self.g.nodes[neighbor]['social']
            else:
                self.g.nodes[person]['social'] -= 0.025*(1-self.g.nodes[neighbor]['social'])
            
            #if a neighbor is not confident, we add a small boost to the individual's own confidence score
            if self.g.nodes[neighbor]['confidence'] < 0.5:
                self.g.nodes[person]['confidence'] += 0.015*(1-self.g.nodes[neighbor]['confidence'])
            else:
                self.g.nodes[person]['confidence'] -= 0.025*self.g.nodes[neighbor]['confidence']
        
        #if the individual's own confidence and social score are above 0.7, the individual's wellness increases.
        if self.g.nodes[person]['confidence'] >= 0.7 and self.g.nodes[person]['social'] >= 0.7:
            self.g.nodes[person]['wellness'] += 0.05*(self.alpha*self.g.nodes[person]['confidence'] + self.beta*self.g.nodes[person]['social'])
        
        #if the individual's own confidence and social score are below 0.4, the individual's wellness decreases.
        elif self.g.nodes[person]['confidence'] <=0.4 and self.g.nodes[person]['social'] <= 0.4:
            self.g.nodes[person]['wellness'] -= 0.05*(self.alpha*self.g.nodes[person]['confidence'] + self.beta*self.g.nodes[person]['social'])
        
        #a node will randomly make a friend with probability = social score
        if np.random.uniform() < self.g.nodes[person]['social']:
            self.g.add_edge(person, np.random.choice(list(self.g)))
            
        #node will randomly unfriend someone if their social score is lower than 0    
        if self.g.nodes[person]['social'] < 0.1 and len(list(self.g.neighbors(person))) > 0:
            self.g.remove_edge(person, np.random.choice(list(self.g.neighbors(person))))
        
```

```python
sim = Wellness(alpha=1.2, beta=1.2)

if interactive:
    # Set up the interactive user interface.
    import pycxsimulator
    pycxsimulator.GUI().start(func=[sim.initialize, sim.observe, sim.update])

else:
    # Run the simulator for 100 steps and show some of the steps.
    print('Initial state:')
    sim.initialize()
    sim.observe()

    #print('Next 5 states:')
    #for i in range(5):
    #    sim.update()
    #    sim.observe()
    #    print('')

    
    for i in range(10):
        print(f'Time-step {i+1} (100 steps):')
        for i in range(100):
            sim.update()
        sim.observe()
    
```
