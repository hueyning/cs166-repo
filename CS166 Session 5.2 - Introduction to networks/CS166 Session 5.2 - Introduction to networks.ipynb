<h3>Sayama Exercise 15.2.</h3>

Calculate the following:

• The number of edges in a complete graph made of n nodes

$n(n-1)/2$

• The number of edges in a regular graph made of n nodes each of which has
degree k

Each node n has k edges coming out from it, so we get $kn/2$ total edges. We divide by 2 since each each edge is connected to 2 nodes, so by taking $kn$ we're overcounting by a factor of 2.

<h3>Sayama Exercise 15.4.</h3>

Every connected tree graph made of n nodes has exactly n  – 1
edges. Explain why.

From the root node down, every node is connected to a parent node by a single edge. As such, all nodes have one "parent" edge except for the root node itself. So with n nodes, there are n-1 edges.

<h3>Bonus Exercise</h3>

Imagine a graph representing hockey players and hockey clubs. Each node is either a
player or a club and each edge indicates that a player has played for that club at some
point in the past. Explain why this graph is a bipartite graph.

<h4>Answer</h4>
The graph would be a bipartite graph since the edge indicates a player having played for a club at some point in the past. As such, we will never see edges between "player" nodes or "club" nodes, so we can divide the nodes into 2 sets that will never be connected to each other: 1 set contains the players, and 1 set contains the clubs.
