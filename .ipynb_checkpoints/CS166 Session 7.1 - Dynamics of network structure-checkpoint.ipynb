<b>Sayama Exercise 16.14</b>

Revise the small­world network formation model
above so that the network is initially a two­dimensional grid in which each node is
connected to its four neighbors. (north, south, east, and west; except for those on the
boundaries of the space). Then run the simulations, and see how random edge rewiring
changes the topology of the network.

```python
import matplotlib.pyplot as plt

# Set this variable to True to run an interactive simulator. This will not work on CoCalc.
interactive = False  
#comment out 'matplotlib inline' if using the GUI
%matplotlib inline 

if interactive:
    matplotlib.use('TkAgg')
    
import pylab
import networkx as nx
import random as rd


class smallWorld:
    
    def __init__(self, interactive):
        self.g = nx.Graph()
        self.n = 30 # number of nodes
        self.k = 4 # number of neighbors of each node
        self.s = 6 # length of grid side
        
        self.interactive=interactive
        
    def initialize(self):
        self.g = nx.Graph()
        for i in range(self.s):
            for j in range(self.s):
                if j < self.s-1:
                    self.g.add_edge(i*self.s+j,i*self.s+j+1)
                if i < self.s-1:
                    self.g.add_edge(i*self.s+j,(i+1)*self.s+j)
                    
        self.g.pos = nx.spring_layout(self.g)
        self.g.count = 0

    def observe(self):
        
        if self.interactive:
            pylab.cla()  # Clear the current figure
        else:
            pylab.figure()  # Create a new figure

        nx.draw(self.g, pos = self.g.pos,
               with_labels=True, cmap="Set3", node_color = [i for i in self.g.nodes])
        
        if not self.interactive:
            pylab.show()

    def update(self):
        self.g.count += 1
        if self.g.count % 20 == 0: # rewiring once in every 20 steps
            nds = list(self.g.nodes)
            i = rd.choice(nds)
            if self.g.degree[i] > 0:
                self.g.remove_edge(i, rd.choice(list(self.g.neighbors(i))))
                nds.remove(i)
                for j in self.g.neighbors(i):
                    nds.remove(j)
                self.g.add_edge(i, rd.choice(nds))

        # simulation of node movement
        self.g.pos = nx.spring_layout(self.g, pos = self.g.pos, iterations = 5)
```

```python
sim = smallWorld(interactive=interactive)

if interactive:
    # Set up the interactive user interface.
    import pycxsimulator
    pycxsimulator.GUI().start(func=[sim.initialize, sim.observe, sim.update])

else:
    # Run the simulator for 100 steps and show some of the steps.
    print('Initial state:')
    sim.initialize()
    sim.observe()

    print('Next 5 states:')
    for i in range(5):
        sim.update()
        sim.observe()
        print('')

    print('After 100 steps:')
    for i in range(95):
        sim.update()
    sim.observe()
```

<b>Sayama Exercise 16.15</b>

Simulate the Barabási­Albert network growth
model with m  = 1 , m  = 3 , and m  = 5 , and see how the growth process may be affected by
the variation of this parameter

```python
class BarabasiAlbert:
    
    def __init__(self, m0=5, m=2, interactive=False):
        self.m0 = m0 # number of nodes in initial condition
        self.m = m # number of edges per new node
        self.g = nx.complete_graph(self.m0)

        self.interactive = interactive
    
    def initialize(self):

        self.g = nx.complete_graph(self.m0)
        self.g.pos = nx.spring_layout(self.g)
        self.g.count = 0

    def observe(self):

        if self.interactive:
            pylab.cla()  # Clear the current figure
        else:
            pylab.figure()  # Create a new figure

        nx.draw(self.g, pos = self.g.pos,
               with_labels=True, cmap="Set3", node_color = [i for i in self.g.nodes])
        
        if not self.interactive:
            pylab.show()


    def pref_select(self,nds):

        r = uniform(0, sum(self.g.degree(i) for i in nds))
        x=0
        for i in nds:
            x += self.g.degree[i]
            if r <= x:
                return i

    def update(self):

        self.g.count += 1
        if self.g.count % 20 == 0: # network growth once in every 20 steps
            nds = list(self.g.nodes)
            newcomer = max(nds) + 1
            for i in range(self.m):
                j = self.pref_select(nds)
                self.g.add_edge(newcomer, j)
                nds.remove(j)
            self.g.pos[newcomer] = (0, 0)

        # simulation of node movement
        self.g.pos = nx.spring_layout(self.g, pos = self.g.pos, iterations = 5)
```

```python
def plot_sim(sim):
    if interactive:
        # Set up the interactive user interface.
        import pycxsimulator
        pycxsimulator.GUI().start(func=[sim.initialize, sim.observe, sim.update])

    else:
        # Run the simulator for 100 steps and show some of the steps.
        print('Initial state:')
        sim.initialize()
        sim.observe()

        print('Next 5 states:')
        for i in range(5):
            sim.update()
            sim.observe()
            print('')

        print('After 100 steps:')
        for i in range(95):
            sim.update()
        sim.observe()
        
sim = BarabasiAlbert(m=1)
plot_sim(sim)
```

```python
sim = BarabasiAlbert(m=3)
plot_sim(sim)
```

```python
sim = BarabasiAlbert(m=5)
plot_sim(sim)
```
