<h4>1. Experimentally confirm the theoretical threshold for when an epidemic will persist in a
random network — equation (18.28) in the textbook.</h4>

$p_i > \frac{p_r}{(n-1)p_e} = \frac{p_r}{<k>}$

The probability of infection has to be greater than the probability of recovery divided by the average degree in order for an epidemic to persist in the random network.

<h4>Question: Why does using synchronous or asynchronous updating make a
difference?</h4>

With asynchronous update, we may miss out nodes since we are randomly choosing nodes to apply the update method on. With synchronous update, every node will go through the update method. We need to take this into account when observing the evolution of the network as well by adjusting the time-step.

<h4>Question: For the mean field approximation described in Section 18.5, why is it
appropriate to use the synchronous update method and not the asynchronous
one?</h4>

We use the synchronous update method to make the MFA more similar to the approximation that we applied to cellullar Automata (CA).

```python
import matplotlib.pyplot as plt

# Set this variable to True to run an interactive simulator. This will not work on CoCalc.
interactive = False  
#comment out 'matplotlib inline' if using the GUI
%matplotlib inline 

if interactive:
    matplotlib.use('TkAgg')
    
import pylab
import networkx as nx
import random as rd

class SIS:
    
    def __init__(self,p_i=0.5,p_r=0.5,interactive=False):
        
        self.interactive = interactive
        self.g = nx.karate_club_graph()
        self.p_i = p_i # infection probability
        self.p_r = p_r # recovery probability
        
    
    def initialize(self):
        self.g = nx.karate_club_graph()
        self.g.pos = nx.spring_layout(self.g)
        for i in self.g.nodes:
            self.g.nodes[i]['state'] = 1 if rd.random() < .5 else 0

    def observe(self):
         
        if self.interactive:
            pylab.cla()  # Clear the current figure
        else:
            pylab.figure()  # Create a new figure
            
        nx.draw(self.g, vmin = 0, vmax = 1,
                node_color = [self.g.nodes[i]['state'] for i in self.g.nodes],
                labels = nx.get_node_attributes(self.g, 'state'),
                pos = self.g.pos)
        
        if not self.interactive:
            pylab.show()

    def update(self):
        #a = rd.choice(list(self.g.nodes))
        #use for loop instead of random choice for synchronous updating
        for a in list(self.g.nodes):
            if self.g.nodes[a]['state'] == 0: # if susceptible
                b = rd.choice(list(self.g.neighbors(a)))
                if self.g.nodes[b]['state'] == 1: # if neighbor b is infected
                    self.g.nodes[a]['state'] = 1 if rd.random() < self.p_i else 0
            else: # if infected
                self.g.nodes[a]['state'] = 0 if rd.random() < self.p_r else 1
```

```python
sim = SIS(p_i=0.9,p_r=0.5)

if interactive:
    # Set up the interactive user interface.
    import pycxsimulator
    pycxsimulator.GUI().start(func=[sim.initialize, sim.observe, sim.update])

else:
    print('Initial state:')
    sim.initialize()
    sim.observe()

    for i in range(5):
        print(f'Time-step : {i}')
        for j in range(1):
            sim.update()
        sim.observe()
```

<h4>Question: How does the average degree of neighbors (the number of friends of
your friends) compare to the average degree of the graph (your number of
friends)?</h4>

The average degree of neighbors is greater than the average degree of the graph for all 3 graphs (Erdos-Renyi, Barbarasi-Albert, Watts-strogatz), but particularly so for the Barbarasi-albert graph.

```python
import numpy as np

def get_avg_degree(g):
    '''
    returns average degree of graph
    '''
    return np.mean(list(dict(g.degree()).values()))

def get_avg_neighbor_degree(g):
    '''
    returns average neighbor degree of graph
    '''
    return np.mean(list(nx.average_neighbor_degree(g).values()))
    

#number of nodes
n = 1000
```

```python
#erdos-renyi graph
er = nx.erdos_renyi_graph(n, p=0.04)
nx.draw(er)
get_avg_degree(er)
```

```python
get_avg_neighbor_degree(er)
```

```python
#barabasi-albert grpah
ba = nx.barabasi_albert_graph(n, 21)
nx.draw(ba)
get_avg_degree(ba)
```

```python
get_avg_neighbor_degree(ba)
```

```python
#watts-strogatz graph
ws = nx.watts_strogatz_graph(n, 40, p=0.5)
nx.draw(ws)
get_avg_degree(ws)
```

```python
get_avg_neighbor_degree(ws)
```
