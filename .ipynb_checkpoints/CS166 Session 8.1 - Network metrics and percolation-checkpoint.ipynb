<h2>Percolation</h2>

Following the mathematical analysis in the textbook, n is defined as the number of nodes in a
network, and $q$ is defined as the probability that a random node from the network is not part of
the largest connected component (LCC) of the network.

The derivation in the textbook shows that $q$ is a solution to the equation

$q = e^{⟨k⟩ (q−1)}$       (1) 

where $⟨k⟩$ is the average degree of the network.

<b>Question: Given the information above, what is the theoretical estimate for the number of
nodes in the LCC, expressed in terms of the known variables, $n, q$, and $⟨k⟩$ ?</b>

The theoretical estimate for the number of nodes in the LCC, $p$, is:

$p = n(1-q) = n(1-e^{⟨k⟩ (q−1)})$

<b>Task: Plot how the size of the LCC depends on the average degree $⟨k⟩$ by using the theoretical
result in (1).</b>


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()

def calculate_q(k):
    '''
    Use a numerical root finder to determine q from the equation
    q = exp(k*(q-1)).
    '''
    from scipy.optimize import root
    return root(lambda q: q - np.exp(k * (q - 1)), 0).x[0]

n_vals = [10**i for i in range(6)]
k_vals = np.arange(1,10,0.5)

plt.figure(figsize=(25,12))
for n in range(len(n_vals)):
    plt.subplot(2,3,n+1)
    p_vals = [n_vals[n]*(1-calculate_q(k)) for k in k_vals]
    plt.title(f"Average degree vs Size of LCC (Graph size: {n_vals[n]})")
    plt.ylabel("Size of LCC")
    plt.xlabel("Average degree")
    plt.plot(k_vals, p_vals)
```
